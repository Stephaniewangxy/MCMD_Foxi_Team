**<font>Pythagorean Expected Winning Percentage<font>**

```python
data['exp_win1'] = (data['team1_adjoe']**11.5)/ ((data['team1_adjde']**11.5)+(data['team1_adjoe']**11.5))
data['exp_win2'] = (data['team2_adjoe']**11.5)/ ((data['team2_adjde']**11.5)+(data['team2_adjoe']**11.5))
```

```python
#adjoe:An estimate of the offensive efficiency (points scored per 100 possessions) a team would have against the average D-I defense.
#adjed:An estimate of the defensive efficiency (points allowed per 100 possessions) a team would have against the average D-I offense
```

**<font>Log5 prediction for A's chance of winning<font>**

```python
data['team1_log5'] = (data['exp_win1'] - (data['exp_win1']*data['exp_win2']))/ (data['exp_win1']+data['exp_win2']-(2*data['exp_win1']*data['exp_win2']))
```
